```python
import os
import eelbrain as eel
import pandas as pd
import numpy as np
import pathlib
from pathlib import Path
os.sys.path.append(str(Path().cwd().resolve().parent.parent))
import NLGC_Utils
from NLGC_Utils import LinkAnalyzer
import experiment
import matplotlib.pyplot as plt
```

```python
modeldir = pathlib.Path().cwd() / "../data/nlgc"

models = []
for model in sorted(list(modeldir.glob("*_fullmodel_deltaband.pickle"))):
    print(model.name)
    sub = model.name[5:10]
    if len(model.name) == 44:
        epochno = int(model.name[16])
    else:
        epochno = int(model.name[16:18])
    
    out = eel.load.unpickle(model)
    models.append([sub, epochno, out])
```

```python
exclusions = ["R2303", "R2304", "R2309", "R2306", "R2311"]

lobe_mapping = { # lobe mapping parcellation
    "F": ['rostralmiddlefrontal', 'caudalmiddlefrontal', 'parsopercularis', 
                            'parstriangularis'], 
    "P": ['inferiorparietal', 'posteriorcingulate'], 
    "T": ['superiortemporal', 'middletemporal', 'transversetemporal']
}

df = pd.read_excel("../data/schizodatafiles/summary.xlsx")
subs_conds = df.groupby("Group").apply(lambda x: x.MEG_ID)
ncsubs = subs_conds['NC'].values.tolist()
szsubs = subs_conds['SZ'].values.tolist()

subs = np.unique([i[0] for i in models])
ncsubs = [i for i in ncsubs if i in subs]
szsubs = [i for i in szsubs if i in subs]

# use visit as epoch index; this is a hack but the code is set up to to 
LA = LinkAnalyzer(experiment, ncsubs, szsubs, [''], [''], list(range(24)),
                  lobe_mapping=lobe_mapping, 
                  file_patterns="../data/nlgc/nlgc_{subject}_epoch{trial}_fullmodel_deltaband.pickle",
                  subjects_dir="../data/mri/", exclusions=exclusions)

```

```python
condition1 = []
for sub in ncsubs:
    if sub in exclusions:
        continue
    for i in range(24):
        condition1.append([sub, '', '', f'{i}', [('lh', 'lh'), ('rh', 'rh')]])

condition2 = []
for sub in ncsubs:
    if sub in exclusions:
        continue
    for i in range(24):
        condition2.append([sub, '', '', f'{i}', [('lh', 'rh'), ('rh', 'lh')]])
        
condition3 = []
for sub in szsubs:
    if sub in exclusions:
        continue
    for i in range(24):
        condition3.append([sub, '', '', f'{i}', [('lh', 'lh'), ('rh', 'rh')]])
        
condition4 = []
for sub in szsubs:
    if sub in exclusions:
        continue
    for i in range(24):
        condition4.append([sub, '', '', f'{i}', [('lh', 'rh'), ('rh', 'lh')]])
```

```python
LA.heatmap(condition1, condition2, 
                  'nc', 'sz', hemi=True, overlay_nums=True, vmin=0, vmax=0.1, 
                   diffvmin=-0.05, diffvmax=0.05, figsize=(15, 4))
```
